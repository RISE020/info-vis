### Percentage Growth

As discussed earlier, the growth of Schiphol Airport is a significant issue, drawing attention from the EU, USA, and the Dutch parliament. The year 2005 was chosen as the base year, aligning with the EU's baseline in the 'Paris Agreement,' which sets climate goals for 2050. To achieve climate neutrality, the European Union has mandated that aviation emissions must be reduced by 90% by 2050. However, our graph depicting the yearly growth percentages of carbon dioxide emissions at Schiphol Airport shows a different trend. Since 2005, the base year used by the EU for their 90% reduction target, Schiphol Airport has experienced a 17.2% increase in carbon dioxide emissions, contrary to the desired decrease outlined by the EU and Dutch Parliament.

The chart shows that from 2005 to 2015, CO2 emissions grew significantly, outpacing the growth in flight movements and PM emissions. The COVID-19 pandemic in 2020 caused a sharp decline in all metrics. Post-pandemic recovery in 2021 and 2022 shows a rebound, though not to pre-pandemic levels. Overall, while flight movements have increased, PM emissions have decreased, suggesting improved efficiency or mitigation measures. However, CO2 emissions continue to rise, highlighting the ongoing challenge of reducing greenhouse gas emissions in aviation.
